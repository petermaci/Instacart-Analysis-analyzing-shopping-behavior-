```python
ords_prods_merge['orders_days_of_week'].value_counts(dropna = False)
```

```python
# Updating results and grouping busiests, regular, and least busy days

result = []

for value in ords_prods_merge["orders_days_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 1:
        result.append('Busiest day')
        
        
  elif value == 4:
    result.append("Least busy")
  elif value == 3:
    result.append("Least busy")
    
  else:
    result.append("Regularly busy")
    
```

```python
ords_prods_merge['busiest_day'] = result
```

```python
new_result = []

for value in ords_prods_merge["orders_days_of_week"]:
  if value == 0:
    new_result.append("Busiest_days")
  elif value == 1:
    new_result.append("Busiest_days")
  elif value == 3:
    new_result.append("Slowest_days")
  elif value ==4:
    new_result.append("Slowest_days")
  else:
    new_result.append("Regularly busy")
```

```python
# 4 --- When too many users make Instacart orders at the same time, the app freezes. 
# The senior technical officer at Instacart wants you to identify the busiest hours of the day. 
# Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” 
# Create a new column containing these labels called “busiest_period_of_day.”
```

```python
ords_prods_merge['order_time'].value_counts(dropna = False)
```

```python
# The busiest hours of the day are 10, 11, 14, 15, 13, 12, 16, 9, 16, 8... and so on, with 10 and 11 am being the busiest.
# Essentially, the busiest times for orders are mid-morning to early afternoon.
```

```python
ords_prods_merge.loc[(ords_prods_merge['order_time'].isin([2,3,4,5]), 'busiest_period_of_day')] = 'Fewest Orders'
```

```python
update_result = []

for value in ords_prods_merge["order_time"]:
  if value > 8 & value <=17:
    update_result.append("Most_orders")
  elif value <=5:
    update_result.append("Fewest_orders")
  else:
    update_result.append("Average_orders")
```

```python
update_result
```

```python
ords_prods_merge['Busiest_period_of_day'] = update_result
```
